### 7d

```python
# Datenpunkte (t, b)
t = np.array([0, 1, 2])
b = np.array([1, 2, 2])

# Designmatrix A für b = x1 + x2*t
A = np.column_stack((np.ones_like(t), t))   # [[1,0],[1,1],[1,2]]

# Least Squares Lösung: x̂ = (A^T A)^(-1) A^T b
x_hat = np.linalg.inv(A.T @ A) @ (A.T @ b)
x1_hat, x2_hat = x_hat

print("x_hat =", x_hat)  # erwartet: [7/6, 1/2] ≈ [1.16666667, 0.5]

# Regressionsgerade
t_line = np.linspace(t.min(), t.max(), 200)
b_line = x1_hat + x2_hat * t_line

# Plot
plt.figure()
plt.scatter(t, b, label="Datenpunkte")
plt.plot(t_line, b_line, label=f"Regression: b = {x1_hat:.4f} + {x2_hat:.4f} t")
plt.xlabel("t")
plt.ylabel("b")
plt.title("Datenpunkte und Regressionsgerade")
plt.legend()
plt.grid(True)
plt.show()
```
