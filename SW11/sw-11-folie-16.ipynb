```python
import numpy as np
import numpy.linalg as la
```

```python
A = np.array ([
    [ 4, 4],
    [-3, 3]
])
```

```python
U0, sigma0, V0T = la.svd (A)
V0 = V0T.T
```

```python
U0
```

```python
Sigma0 = np.diag (sigma0)
Sigma0
```

```python
V0
```

```python
U0 @ Sigma0 @ V0.T
```

```python
U0 @ Sigma0 @ V0.T - A
```

```python
EW1, EV1 = la.eig (A @ A.T)
```

```python
U1 = EV1
U1
```

```python
Sigma1 = np.diag (np.sqrt (EW1))
Sigma1
```

```python
V1 = A.T @ U1 @ la.inv (Sigma1)
V1
```

```python
U1 @ Sigma1 @ V1.T
```

```python
U1 @ Sigma1 @ V1.T - A
```

```python
EW2, EV2 = la.eig (A.T @ A)
```

```python
V2 = EV2
V2
```

```python
Sigma2 = np.diag (np.sqrt (EW2))
Sigma2
```

```python
U2 = A @ V2 @ la.inv (Sigma2)
U2
```

```python
U2 @ Sigma2 @ V2.T
```

```python
U2 @ Sigma2 @ V2.T - A
```
