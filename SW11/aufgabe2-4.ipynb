# Aufgabe 2

```python
import numpy as np
import numpy.linalg as la
import math as mt
```

```python
A = np.array ([[1, 2], # Singulär, da beide Zeilen Identisch
               [3, 6]])
```

```python
I = np.array ([[1, 0],
               [0, 1]]) 
```

```python
ATA = A.T @ A 
print(ATA)
```

```python
#Eigenvektoren berechnen
w, eigenVec = la.eig(ATA)
```

```python
# Eigenvektoren normieren
idx = np.argsort(w)[::-1] # nach grösse sortieren
w = w[idx]
eigenVec = eigenVec[:, idx]

v1 = eigenVec[:, 0] / np.linalg.norm(eigenVec[:, 0])
v2 = eigenVec[:, 1] / np.linalg.norm(eigenVec[:, 1])
V  = np.column_stack((v1, v2)) 

print("v1 =", v1)
print("v2 =", v2)
```

```python
Sigma = np.diag(np.sqrt(w))
print(Sigma)
```

```python
sigma1 = np.sqrt(w[0])      # größter Singularwert

u1 = (A @ v1) / sigma1 

u2 = np.array([-u1[1], u1[0]]) # beliebiger normierte Vektor, der orthogonal zu u1 ist.
u2 = u2 / np.linalg.norm(u2)

print("u1 =", u1)
print("u2 =", u2)
```

```python
U = np.column_stack((u1, u2))
print("U =\n", U)
```

```python
result = U @ Sigma @ V.T

print("A =\n", A)
print("Lösung =\n", result)
```

# Aufgabe 3

```python
A = np.array ([[1, 1, 0], # Singulär, da beide Zeilen Identisch
               [0, 1, 1]])
```

```python
ATA = A.T @ A
print(ATA)
```

```python
#Eigenvektoren berechnen
w, eigenVec = la.eig(ATA)
```

```python
# Eigenvektoren normieren
idx = np.argsort(w)[::-1] # nach grösse sortieren
w = w[idx]
eigenVec = eigenVec[:, idx]

v1 = eigenVec[:, 0] / np.linalg.norm(eigenVec[:, 0])
v2 = eigenVec[:, 1] / np.linalg.norm(eigenVec[:, 1])
V  = np.column_stack((v1, v2)) 

print("v1 =", v1)
print("v2 =", v2)
```

```python
sing = np.sqrt(w)           # alle σ_i
Sigma = np.diag(sing[:2])   # 2x2-Matrix für reduzierte SVD
print("Sigma =\n", Sigma)
```

```python
sigma1 = sing[0]
sigma2 = sing[1]

u1 = (A @ v1) / sigma1
u2 = (A @ v2) / sigma2

print("u1 =", u1)
print("u2 =", u2)
```

```python
U = np.column_stack((u1, u2))
print("U =\n", U)
```

```python
result = U @ Sigma @ V.T

print("A =\n", A)
print("Lösung =\n", np.round(result,2))
```

# Aufgabe 3

```python
A = np.array ([[2, 4], 
               [1, 3],
               [0, 0],
               [0, 0]])
```

```python
U, s, Vt = np.linalg.svd(A)   # Standard: „volle“ SVD

Sigma = np.zeros((4, 2))
Sigma[:len(s), :len(s)] = np.diag(s)  # Sigma als 2x3 bauen

result = U @ Sigma @ Vt

print("U =\n", np.round(U, 2))
print("s =", np.round(s, 2))
print("Vt =\n", np.round(Vt, 2))
print("Rekonstruktion =\n", np.round(result, 2))
```
