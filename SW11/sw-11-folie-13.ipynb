```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
A = np.array ([
    [1, 2],
    [2, 4],
    [3, 6]
])
print (A)
```

```python
U, D, VT = np.linalg.svd (A)
Sigma = np.diag (D)
print (U)
print (Sigma)
print (VT)
```

```python
B = np.array ([
    [1, 0, 0, 0, 2],
    [0, 0, 3, 0, 0],
    [0, 0, 0, 0, 0],
    [0, 2, 0, 0, 0]
])
print (B)
```
