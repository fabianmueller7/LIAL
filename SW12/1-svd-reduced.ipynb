```python
import numpy as np
import numpy.linalg as la
```

```python
A = np.array ([
    [ 1,  1],
    [-1, -1]
])
print (A)
```

```python
U, D, VT = la.svd (A)
```

```python
print (U)
```

```python
print (D)
```

```python
Sigma = np.diag (D)
print (Sigma)
```

```python
print (VT)
```

```python
U_r = U [:, 0:1]
print (U_r)
```

```python
Sigma_r = Sigma [0:1, 0:1]
print (Sigma_r)
```

```python
VT_r = VT [0:1, :]
print (VT_r)
```
