```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as pl
```

```python
n = 1000
```

```python
xx = np.linspace (-1, 1, n)
```

```python
yy = np.sin (np.pi * xx)
```

```python
pl.plot (xx, yy)
```

```python
m = 30
A = np.zeros ((n, m))
for k in range (m):
    A [:, k:k+1] = np.array ([xx ** k]) .T
```

```python
c1 = la.inv (A.T @ A) @ A.T @ yy
print (c1)
```

```python
z1 = A @ c1
```

```python
pl.plot (xx, yy, 'blue')
pl.plot (xx, z1, 'red')
```

```python
c2 = la.pinv (A.T @ A) @ A.T @ yy
z2 = A @ c2
```

```python
pl.subplot (1, 2, 1); pl.plot (xx, yy - z1)
pl.subplot (1, 2, 2); pl.plot (xx, yy - z2)
```
