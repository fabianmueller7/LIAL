```python
import numpy as np
import matplotlib.pyplot as pl
import matplotlib.ticker as tk
```

```python
P2 = np.array ([
    [0, 1, 1, 0, 0],
    [0, 0, 1, 1, 0]
])
```

```python
def Zeichne2 (a, p):
    
    q = a @ p

    pl.axis ('equal')
    pl.grid ('both')

    pl.plot (p [0], p [1], 'red', linewidth = 3)
    pl.plot (q [0], q [1], 'blue')

    ax = pl.gca()
    ax.xaxis.set_major_locator (tk.MultipleLocator (1))
    ax.yaxis.set_major_locator (tk.MultipleLocator (1))
```

# Aufgabe 4

```python
A4 = np.array ([
    [2, 0],
    [0, 2]
])

Zeichne2 (A4, P2)
```

# Aufgabe 5

```python
A5 = np.array ([
    [1, 0],
    [0, 2]
])

Zeichne2 (A5, P2)
```
