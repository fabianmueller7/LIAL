```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as pl

%matplotlib widget
```

```python
def CS (size = (10, 10), xrange = [0, 10], yrange = [0, 10], zrange = [0, 10]):
    fig = pl.figure (figsize = size)
    ax = fig.add_subplot (projection = '3d')
    ax.set_xlim (xrange); ax.set_xlabel ('x')
    ax.set_ylim (yrange); ax.set_ylabel ('y')
    ax.set_zlim (zrange); ax.set_zlabel ('z')
    ax.set_aspect ('equal')
```

```python
def Vec (x, y, z):
    return np.array ([[x], [y], [z]])
```

```python
O = Vec (0, 0, 0)
```

```python
def PlotVec (p, v, col = 'black'):
    pl.quiver (p [0], p [1], p [2], v [0], v [1], v [2], color = col, arrow_length_ratio = 0.1)
```

```python
def PlotVecO (v, col = 'black'):
    PlotVec (O, v, col)
```

```python
u = Vec (1, 2, 3)
v = Vec (4, 5, 6)
w = Vec (-3, 6, -3)
w2 = Vec (3, -6, 3)
```

```python
CS ((5, 5), [-10, 10], [-10, 10], [-10, 10])
PlotVecO (u, 'blue')
PlotVecO (v, 'magenta')
PlotVecO (w, 'red')
PlotVecO (w2, 'green')
pl.show()
```
