```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def Vec (x, y, z):
    return np.array ([[x], [y], [z]])
```

```python
A = Vec (0, -1, 0)
```

```python
B  = Vec (1, 0, 0)
```

```python
C = Vec (0, 1, 0)
```

```python
D = Vec (-1, 0, 0)
```

```python
E = Vec (0, 0, 1)
```

```python
def Cross (u, v):
    return Vec (
        u [1] [0] * v [2] [0] - u [2] [0] * v [1] [0],
        u [2] [0] * v [0] [0] - u [0] [0] * v [2] [0],
        u [0] [0] * v [1] [0] - u [1] [0] * v [0] [0]
    )   
```

```python
def Normal (a, b, c):
    return Cross (b - a, c - a)
```

```python
def Dot (u, v):
    return (u.T @ v) [0] [0]
```

```python
def Norm (u):
    return np.sqrt (Dot (u, u))
```

```python
def Normalize (u):
    return u / Norm (u)
```

```python
def Unitnormal (a, b, c):
    return Normalize (Normal (a, b, c))
```

```python
Light = Normalize (Vec (-1, 1, 2)) # Vec(X-Achse, Y-Achse, Z-Achse)
print (Light)
```

```python
def Plot (a, b, c):
    p = np.concatenate ((a, b, c, a), axis = 1)
    n = Unitnormal (a, b, c)
    col = Dot (n, Light)
    if col < 0:
        col = 0
    if col > 1:
        col = 1
    pl.fill (p [0], p [1], color = (col, col, col))
```

```python
pl.axis ('equal')
Plot (A, B, E)
Plot (B, C, E)
Plot (C, D, E)
Plot (A, E, D)
```
