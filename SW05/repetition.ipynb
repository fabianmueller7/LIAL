```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
P = np.array ([  # Grosses blaues viereck 
    [0, 1, 1, 0, 0],  # X-Achse        Hier werden Punkte erstellt (0,0) -> (2,0) -> (2,2)
    [0, 0, 1, 1 , 0]  # Y-Achse
])

A = np.array ([ #Rotes viereck
    [2, 0],   # X-Achse 
    [0, 2]    # Y-Achse
])
Q = A @ P 

pl.axis ('equal')
pl.plot (P [0], P [1])
pl.plot (Q [0], Q [1], 'red')
```
