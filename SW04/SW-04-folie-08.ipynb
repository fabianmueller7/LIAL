```python
import numpy as np
```

```python
A = np.array ([
    [6, -1, -1,  4],
    [1,  1, 10, -6],
    [2, -1,  1, -2]
])
print (A)
```

```python
np.linalg.solve (A [0:3, 0:3], A [0:3, 3:4])
```
