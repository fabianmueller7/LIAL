```python
import numpy as np
```

```python
A = np.array ([
    [1, 1, 1, 0],
    [4, 2, 1, 1],
    [9, 3, 1, 3.]
])
print (A)
```

```python
A [1] = A [1] - 4 * A [0]
print (A)
```

```python
A [2] = A [2] - 9 * A [0]
print (A)
```

```python
A [2] = A [2] - 3 * A [1]
print (A)
```

```python
A [1] = A [1] + 3 * A [2]
print (A)
```

```python
A [0] = A [0] - A [2]
print (A)
```

```python
A [1] = A [1] / -2
print (A)
```

```python
A [0] = A [0] - A [1]
print (A)
```
