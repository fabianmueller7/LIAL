```python
import numpy as np
import numpy.linalg as la
```

```python
A = np.array ([
    [ 1, 2, 1],
    [ 2, 1, 0],
    [-3, 0, 9.]
])
A
```

```python
E21 = np.array ([
    [ 1, 0, 0],
    [-2, 1, 0],
    [ 0, 0, 1.]
])
E21
```

```python
E21 @ A
```

```python
E31 = np.array ([
    [1, 0, 0],
    [0, 1, 0],
    [3, 0, 1.]
])
E31 @ E21 @ A
```

```python
E32 = np.array ([
    [1, 0, 0],
    [0, 1, 0],
    [0, 2, 1.]
])
R = E32 @ E31 @ E21 @ A
R
```

```python
E21
```

```python
la.inv (E21)
```

```python
la.inv (E31)
```

```python
la.inv (E32)
```

```python
L = la.inv (E21) @ la.inv (E31) @ la.inv (E32)
L
```

```python
L @ R
```

```python
A
```

```python
L @ R - A
```

```python
np.max (np.abs (L @ R - A))
```
