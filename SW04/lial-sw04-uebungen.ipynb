```python
import numpy as np
import numpy.linalg as la
```

## Aufgabe 1

### A

```python
A = np.array ([[4, 3], [8, 6]])
print (la.det (A))
```

### B

```python
B = np.array ([[4, 3], [8, 7]])
print (la.det (B))
```

```python
B_inv = la.inv(B)
print(B_inv)
```

### C

```python
C = np.array ([[6, 6], [6, 6]])
print (la.det (C))
```

### D

```python
D = np.array ([[1, 0, 0], [ 1, 1, 0], [ 1, 1, 1]])
print (la.det (D))
```

```python
D_inv = la.inv(D)
print(D_inv)
```
