```python
import numpy as np
import numpy.linalg as la
import math as mt
```

# Aufgabe 3

### A

```python
A = np.array([
    [1, mt.sqrt(2), -1],
    [-mt.sqrt(2), 0, -mt.sqrt(2)],
    [-1, mt.sqrt(2), 1]
])

A = A*0.5

A_T = A.T #get transpond-matrix

A_I = la.inv(A) #get inverse-matrix


print(A_T)
print("-----------------") # orthogonal, da gleich
print(A_I)
```

### B

```python
A = np.array([
    [1, mt.sqrt(2), -1],
    [-mt.sqrt(2), 0, -mt.sqrt(2)],
    [-1, mt.sqrt(2), 1]
])

A = A*0.25

A_T = A.T #get transpond-matrix

A_I = la.inv(A) #get inverse-matrix


print(A_T)
print("-----------------") # nicht orthogonal, da nicht gleich
print(A_I) 
```
