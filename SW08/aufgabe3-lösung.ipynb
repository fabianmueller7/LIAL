```python
import numpy as np
import numpy.linalg as la
import math as mt
```

```python
X = np.array([
    [1, 1, 1],
    [1, 2, 1],
    [1, 2, 2],
    [1, 1, 2]
])
```

```python
y = np.array([[1], [2], [3], [3]])
```

### b

```python
result = X.T @ X
print(result)
```

### C

```python
b = la.pinv(X.T @ X) @ X.T @ y
print(b)
```

### D

```python
I = np.eye(X.shape[0]) #Einheitsmatrix
A = np.ones((4, 4)) # Matrix gefÃ¼llt mit 1
```

```python
P = X @ la.pinv(X.T @ X) @ X.T
print("P", P)

Q = I - P
print("Q", Q)

M = I - (1/4) * A
print("M", M)
```

### E

```python
y_dach = P @ y
print(y_dach)

e1 = Q @ y
print("e1", e1)

e2 = y - y_dach
print("e2", e1)
```

### F

```python
SQR = float(y.T @ Q @ y)    
SQT = float(y.T @ M @ y)
R2 = 1 - SQR / SQT
print(mt.sqrt(R2))
```
