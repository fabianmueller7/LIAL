```python
import numpy as np
import numpy.linalg as la
```

```python
p = np.array ([
    [3],
    [5],
    [4]
])
```

```python
u = np.array ([
    [4],
    [6],
    [4]
])
```

```python
q = np.array ([
    [8],
    [2],
    [7]
])
```

```python
v = np.array ([
    [3],
    [1],
    [5]
])
```

```python
A3 = np.concatenate ((u, -v), axis = 1)
print (A3)
```

```python
A = A3 [0:2, 0:2]
print (A)
```

```python
b3 = q - p
print (b3)
```

```python
b = b3 [0:2, 0:1]
print (b)
```

```python
[s, t] = la.solve (A, b)
print (s)
print (t)
```

```python
p + s * u
```

```python
q + t * v
```
