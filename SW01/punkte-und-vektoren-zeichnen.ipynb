```python
def PlotVec (p, u, c):
    pl.arrow (p [0] [0], p [1] [0], u [0] [0], u [1] [0],
        color = c,
        head_width = 1,
        length_includes_head = True,
        fill = False)
```

```python
a = np.array ([[-7], [-3]])
a
```

```python
b = np.array ([[13], [-5]])
b
```

```python
c = np.array ([[2], [11]])
c
```

```python
ab = b - a
ab
```

```python
bc = c - b
bc
```

```python
ca = a - c
ca
```

```python
aa = 3 * a
aa
```

```python
o = np.array ([[0], [0]])
```

```python
PlotVec (a, ab, 'blue')
PlotVec (b, bc, 'magenta')
PlotVec (c, ca, 'red')
PlotVec (o, a, 'green')
PlotVec (o, aa, 'green')
pl.gca().set_aspect('equal')
```
