```python
import numpy as np
import numpy.linalg as la
```

```python
p = np.array ([
    [2],
    [3],
    [4.]
])
p
```

```python
q = np.array ([
    [5],
    [3],
    [4.]
])
q
```

```python
u = np.array ([
    [1],
    [2],
    [5.]
])
u
```

```python
v = np.array ([
    [3],
    [4],
    [4.]
])
v
```

```python
w = np.array ([
    [7],
    [3],
    [3.]
])
w
```

```python
A = np.hstack ((u, v, -w))
A
```

```python
b = q - p
b
```

```python
x = la.solve (A, b)
x
```

```python
[r, s, t] = x
print (r, s, t)
```

```python
p + r * u + s * v
```

```python
q + t * w
```
