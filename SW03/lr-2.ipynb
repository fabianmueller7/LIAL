# LR-Zerlegung

```python
import numpy as np
import numpy.linalg as la
import numpy.random as rd
```

## Gleichungssystem lösen

$\begin{align*}
A &\;=\; L \, R \\[6pt]
A \, x &\;=\; b \\
L \, R \, x &\;=\; b \\[6pt]
R \, x &\;=\; y \\
L \, y &\;=\; b
\end{align*}$

## Testdaten

```python
rd.seed (1)
```

```python
def RandMat (m, n):
    max = 10
    min = - max
    return np.round (rd.rand (m, n) * (max - min) + min, 0)
```

```python
def pr (txt, mat):
    fmt = '%10.5f'
    spc = ' ' * len (txt)
    m, n = mat.shape
    for i in range (m):
        print (txt if i == 0 else spc, end = '')
        for k in range (n):
            print ('', fmt % mat [i, k], end = '')
        print()

def Pr (txt, mat):
    pr (txt, mat)
    print()
```

```python
class Test:
    n = 5
    A = RandMat (n, n)
    x = RandMat (n, 1)
    b = A @ x

Pr ('A:               ', Test.A)
Pr ('b:               ', Test.b.T)
pr ('x:               ', Test.x.T)
pr ('la.solve (A, b): ', la.solve (Test.A, Test.b) .T)
```

## LR-Zerlegung mit Eliminationsmatrizen

```python
def LRmat (A, b):
    
    n = A.shape [0]
    L = np.eye (n)
    R = np.copy (Test.A)

    Pr ('L ', L)
    pr ('R ', R)

    for i in range (n - 1):
        for k in range (i + 1, n):
            E = np.eye (n); E [k, i] = - R [k, i] / R [i, i]
            R = E @ R
            L = L @ la.inv (E)
            print ('--------------------', i, k, '--------------------')
            Pr ('E ', E)
            Pr ('L ', L)
            pr ('R ', R)

    print ('----------------------------------------')
    Pr ('A ', A)
    Pr ('L ', L)
    Pr ('R ', R)
    pr ('LR', L @ R)

    print ('----------------------------------------')
    y = la.solve (L, b)
    x = la.solve (R, y)
    Pr ('y      ', y.T)
    pr ('x      ', x.T)
    pr ('Test.x ', Test.x.T)

LRmat (Test.A, Test.b)
```

## LR-Zerlegung ausprogrammiert

### Zerlegung (ausprogrammiert)

```python
def LRdecomp1 (A):

    n = A.shape [0]
    L = np.zeros ((n, n))
    R = np.copy (A)

    for i in range (n):
        L [i, i] = 1
        for k in range (i + 1, n):
            t = R [k, i] / R [i, i]
            L [k, i] = t
            for j in range (i, n):
                R [k, j] -= t * R [i, j]

    return L, R
```

```python
def TestLRdecomp1 (A):

    L, R = LRdecomp1 (A)

    Pr ('A ', A)
    Pr ('L ', L)
    Pr ('R ', R)
    pr ('LR', L @ R)

TestLRdecomp1 (Test.A)
```

### Lösung (ausprogrammiert)

```python
def LRsolve1 (L, R, b):
    
    n = L.shape [0]

    y = np.empty ((n, 1))
    for i in range (n):
        t = b [i, 0]
        for k in range (0, i):
            t -= L [i, k] * y [k, 0]
        y [i, 0] = t

    x = np.empty ((n, 1))
    for i in range (n - 1, -1, -1):
        t = y [i, 0]
        for k in range (i + 1, n):
            t -= R [i, k] * x [k, 0]
        x [i, 0] = t / R [i, i]

    return x
```

```python
def TestLRsolve1 (A, b):

    L, R = LRdecomp1 (A)
    x = LRsolve1 (L, R, b)

    pr ('x      ', x.T)
    pr ('Test.x ', Test.x.T)

TestLRsolve1 (Test.A, Test.b)
```

## LR-Zerlegung speichereffizient

### Zerlegung (speichereffizient)

```python
def LRdecomp2 (A):

    n = A.shape [0]
    LR = np.copy (A)

    for i in range (n):
        for k in range (i + 1, n):
            t = LR [k, i] / LR [i, i]
            LR [k, i] = t
            for j in range (i + 1, n):
                LR [k, j] -= t * LR [i, j]

    return LR
```

```python
def TestLRdecomp2 (A):

    LR = LRdecomp2 (A)
    L = np.tril (LR, -1) + np.eye (LR.shape [0])
    R = np.triu (LR)

    Pr ('Res', LR)
    Pr ('L  ', L)
    Pr ('R  ', R)
    pr ('LR ', L @ R)

TestLRdecomp2 (Test.A)
```

### Lösung (speichereffizient)

```python
def LRsolve2 (LR, b):
    
    n = LR.shape [0]

    y = np.empty ((n, 1))
    for i in range (n):
        t = b [i, 0]
        for k in range (0, i):
            t -= LR [i, k] * y [k, 0]
        y [i, 0] = t

    x = np.empty ((n, 1))
    for i in range (n - 1, -1, -1):
        t = y [i, 0]
        for k in range (i + 1, n):
            t -= LR [i, k] * x [k, 0]
        x [i, 0] = t / LR [i, i]

    return x
```

```python
def TestLRsolve2 (A, b):

    LR = LRdecomp2 (A)
    x = LRsolve2 (LR, b)

    Pr ('x ', x.T)
    pr ('Ax', (A @ x) .T)
    pr ('b ', b.T)

TestLRsolve2 (Test.A, Test.b)
```

## LR-Zerlegung mit Pivotisierung

### Zerlegung (mit Pivotisierung)

```python
def LRdecomp3 (A):

    n = A.shape [0]
    LR = np.copy (A)
    perm = np.arange (n)

    for i in range (n):
        j = i
        for k in range (i + 1, n):
            if (abs (LR [k, i]) > abs (LR [j, i])):
                j = k
        if i != j:
            for k in range (n):
                t = LR [i, k]; LR [i, k] = LR [j, k]; LR [j, k] = t
            t = perm [i]; perm [i] = perm [j]; perm [j] = t
        for k in range (i + 1, n):
            t = LR [k, i] / LR [i, i]
            LR [k, i] = t
            for j in range (i + 1, n):
                LR [k, j] -= t * LR [i, j]

    return LR, perm
```

```python
def TestLRdecomp3 (A):
    
    LR, perm = LRdecomp3 (A)
    L = np.tril (LR, -1) + np.eye (LR.shape [0])
    R = np.triu (LR)

    Pr ('Result', LR)
    
    print ('Permutation:', perm, '\n')
    
    Pr ('L     ', L)
    Pr ('R     ', R)
    Pr ('LR    ', L @ R)
    pr ('A     ', A)

TestLRdecomp3 (Test.A)
```

### Lösung (mit Pivotisierung)

```python
def LRsolve3 (LR, perm, b):
    
    n = LR.shape [0]

    y = np.empty ((n, 1))
    for i in range (n):
        t = b [perm [i], 0]
        for k in range (0, i):
            t -= LR [i, k] * y [k, 0]
        y [i, 0] = t

    x = np.empty ((n, 1))
    for i in range (n - 1, -1, -1):
        t = y [i, 0]
        for k in range (i + 1, n):
            t -= LR [i, k] * x [k, 0]
        x [i, 0] = t / LR [i, i]

    return x
```

```python
def TestLRsolve3 (A, b):

    LR, perm = LRdecomp3 (A)
    x = LRsolve3 (LR, perm, b)

    Pr ('x ', x.T)
    pr ('Ax', (A @ x) .T)
    pr ('b ', b.T)

TestLRsolve3 (Test.A, Test.b)
```
