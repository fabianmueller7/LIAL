```python
import numpy as np
```

```python
A = np.array ([
    [ 1,  1, 2],
    [-1, -2, 3],
    [ 3, -7, 4.]
])
print (A)
```

```python
b = np.array ([
    [8],
    [1],
    [10.]
])
print (b)
```

```python
E21 = np.array ([
    [1, 0, 0],
    [1, 1, 0],
    [0, 0, 1.]
])
print (E21)
```

```python
print (E21 @ A)
```

```python
E31 = np.array ([
    [ 1, 0, 0],
    [ 0, 1, 0],
    [-3, 0, 1.]
])
print (E31)
```

```python
print (E31 @ E21 @ A)
```

```python
E32 = np.array ([
    [1,   0, 0],
    [0,   1, 0],
    [0, -10, 1.]
])
print (E32)
```

```python
print (np.round (E32 @ E31 @ E21 @ A))
```

```python
U = E32 @ E31 @ E21 @ A
print (np.round (U, 10))
```

```python
L = np.linalg.inv (E32 @ E31 @ E21)
print (np.round (L, 10))
```

```python
print (L @ U)
```

```python
print (np.round (L @ U - A, 10))
```

```python
y = np.linalg.inv (L) @ b
print (y)
```

```python
x = np.linalg.inv (U) @ y
print (x)
```

```python
print (A @ x)
```

```python
print (A @ x - b)
```
