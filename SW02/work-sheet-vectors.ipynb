## Predefined Routines

### Libraries

```python
# import libraries

import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as pl
import matplotlib.ticker as tk
```

### Vectors

#### define vector

```python
# Vec (x, y)

def Vec (x, y):
    return np.array ([[x], [y]])
```

#### zero vector

```python
# O

O = Vec (0, 0)
```

_Hint: Capital letter O, not number 0!_

#### dot product

```python
# Dot (vector1, vector2)

def Dot (u, v):
    uv = u.T @ v # @ ist ein Matrixmulitplikator
    return uv [0, 0]
```

#### norm

```python
# Norm (vector)

def Norm (u):
    return np.sqrt (Dot (u, u))
```

### Angles

#### convert degrees to radians

```python
# Rad (deg)

def Rad (deg):
    return deg * np.pi / 180
```

#### convert radians to degrees

```python
# Deg (rad)

def Deg (rad):
    return rad * 180 / np.pi
```

### Plot Routines

#### plot co-ordinate system

```python
# PlotCS()

def PlotCS (size = (5, 4), xticks = (1, 0.5), yticks = (1, 0.5)):

    fig = pl.figure (figsize = size)
    fig.canvas.toolbar_visible = False
    fig.canvas.header_visible = False
    fig.canvas.footer_visible = False

    ks = fig.add_subplot()
    ks.axis ('equal')
    ks.grid (which = 'both', axis = 'both')
    ks.xaxis.set_major_locator (tk.MultipleLocator (xticks [0]))
    ks.xaxis.set_minor_locator (tk.MultipleLocator (xticks [1]))
    ks.yaxis.set_major_locator (tk.MultipleLocator (yticks [0]))
    ks.yaxis.set_minor_locator (tk.MultipleLocator (yticks [1]))

    return ks
```

#### plot vector

##### at given point

```python
# PlotVec (point, direction [, color])

VecHeadWidth = 0.15

def PlotVec (p, u, col = 'black'):
    pl.arrow (p [0] [0], p [1] [0], u [0] [0], u [1] [0],
        color = col,
        head_width = VecHeadWidth,
        length_includes_head = True,
        fill = True,
        zorder = 10)
```

##### at origin

```python
# PlotVecO (direction [, color]): 

def PlotVecO (u, col = 'black'):
    PlotVec (O, u, col)
```

_Hint: Capital letter O, not number 0!_

#### plot circle

##### with center and radius

```python
# PlotCircle (center, radius, [, color]

def PlotCircle (c, r, col = 'black'):
    tt = np.linspace (0, 2 * np.pi, 360)
    pp = c + r * np.array ([np.cos (tt), np.sin (tt)])
    pl.plot (pp [0], pp [1], col)
```

##### unit circle

```python
# plot unit Circle: PlotUnitCircle ([, color])

def PlotUnitCircle (col = 'black'):
    PlotCircle (O, 1, col)
```
