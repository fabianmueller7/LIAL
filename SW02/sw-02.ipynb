```python
b = np.array ([
    [7],
    [1] 
])
```

```python
p = np.array ([
    [1],
    [0],
    [2]
])
```

```python
u = np.array ([
    [1],
    [2],
    [1]
])
```

```python
q = np.array ([
    [2],
    [3],
    [1]
])
```

```python
v = np.array ([
    [4],
    [2],
    [9]
])
```

```python
w = np.array ([
    [3],
    [1],
    [6]
])
```

```python
A = np.hstack ((u, -v, -w))
A
```

```python
b = q - p
b
```

```python
x = la.solve (A, b)
x
```
